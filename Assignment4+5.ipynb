```python
# read the CDS file
cdsdata = pd.read_csv('cds_spread.csv')

cdsdata['year'] = pd.DatetimeIndex(cdsdata['mdate']).year
cdsdata['month'] = pd.DatetimeIndex(cdsdata['mdate']).month
cdsdata.head(5)
```

```python
import numpy as np
CRSPdata = pd.read_csv('Quarterly Merged CRSP-Compustat.csv')
CRSPdata.columns = map(str.lower, CRSPdata.columns)


CRSPdata['year'] = pd.DatetimeIndex(CRSPdata['datadate']).year
CRSPdata['month'] = pd.DatetimeIndex(CRSPdata['datadate']).month

CRSPdata['quarter'] = np.where( CRSPdata['month'].isin([1,2,3]), 'Quarter1', '')
CRSPdata['quarter'] = np.where( CRSPdata['month'].isin([4,5,6]), 'Quarter2', CRSPdata['quarter'])
CRSPdata['quarter'] = np.where( CRSPdata['month'].isin([7,8,9]), 'Quarter3', CRSPdata['quarter'])
CRSPdata['quarter'] = np.where( CRSPdata['month'].isin([10,11,12]), 'Quarter4', CRSPdata['quarter'])
CRSPdata['match_period'] = CRSPdata['year'].astype(str) + CRSPdata['quarter'].astype(str)

CRSPdata.head(5)
```

```python
data1 = CRSPdata[['gvkey','month']]
data1 = data1.drop_duplicates(subset='gvkey', keep='first')

s1 = data1.loc[data1['month'].isin([1,4,7,10])]
s2 = data1.loc[data1['month'].isin([2,5,8,11])]
s3 = data1.loc[data1['month'].isin([3,6,9,12])]

t1 = s1['gvkey'].tolist()
t2 = s2['gvkey'].tolist()
t3 = s3['gvkey'].tolist()
t4 = CRSPdata['gvkey'].unique().tolist()
```

```python
cdsdata1 = pd.read_csv('cds_spread.csv')

cdsdata1['year'] = pd.DatetimeIndex(cdsdata1['mdate']).year
cdsdata1['month'] = pd.DatetimeIndex(cdsdata1['mdate']).month
cdsdata1.head(5)
```

```python
df2 = cdsdata[cdsdata['gvkey'].isin(t1)].reset_index() 
df3 = cdsdata[cdsdata['gvkey'].isin(t2)].reset_index() 
df4 = cdsdata[cdsdata['gvkey'].isin(t3)].reset_index() 
df5 = cdsdata[~cdsdata['gvkey'].isin(t4)].reset_index() 
df2['quarter'] = np.where( df2['month'].isin([1,2,3]), 'Quarter1', '')
df2['quarter'] = np.where( df2['month'].isin([4,5,6]), 'Quarter2', df2['quarter'])
df2['quarter'] = np.where( df2['month'].isin([7,8,9]), 'Quarter3', df2['quarter'])
df2['quarter'] = np.where( df2['month'].isin([10,11,12]), 'Quarter4', df2['quarter'])
df2['match_period'] = df2['year'].astype(str) + df2['quarter']
df2_merged = pd.merge(df2, CRSPdata, on=['gvkey','match_period'], how='left')
df2_merged.head(3)
```

```python
df3['quarter'] = np.where( df3['month'].isin([1,2,3]), 'Quarter1', '')
df3['quarter'] = np.where( df3['month'].isin([4,5,6]), 'Quarter2', df3['quarter'])
df3['quarter'] = np.where( df3['month'].isin([7,8,9]), 'Quarter3', df3['quarter'])
df3['quarter'] = np.where( df3['month'].isin([10,11,12]), 'Quarter4', df3['quarter'])
df3['match_period'] = df3['year'].astype(str) + df3['quarter']
df3_merged = pd.merge(df3, CRSPdata, on=['gvkey','match_period'], how='left')
df3_merged.head(3)
```

```python
df4['quarter'] = np.where( df4['month'].isin([1,2,3]), 'Quarter1', '')
df4['quarter'] = np.where( df4['month'].isin([4,5,6]), 'Quarter2', df4['quarter'])
df4['quarter'] = np.where( df4['month'].isin([7,8,9]), 'Quarter3', df4['quarter'])
df4['quarter'] = np.where( df4['month'].isin([10,11,12]), 'Quarter4', df4['quarter'])
df4['match_period'] = df4['year'].astype(str) + df4['quarter']
df4_merged = pd.merge(df4, CRSPdata, on=['gvkey','match_period'], how='left')
df4_merged.head(3)
```

```python
df5['quarter'] = np.where( df5['month'].isin([1,2,3]), 'Quarter1', '')
df5['quarter'] = np.where( df5['month'].isin([4,5,6]), 'Quarter2', df5['quarter'])
df5['quarter'] = np.where( df5['month'].isin([7,8,9]), 'Quarter3', df5['quarter'])
df5['quarter'] = np.where( df5['month'].isin([10,11,12]), 'Quarter4', df5['quarter'])
df5['match_period'] = df5['year'].astype(str) + df5['quarter']
df5_merged = pd.merge(df5, CRSPdata, on=['gvkey','match_period'], how='left')
df5_merged.head(3)
```

```python
data2 = pd.concat([df2_merged, df3_merged, df4_merged, df5_merged], ignore_index=True)

data2 = data2.drop(['index','Unnamed: 0','year_y','month_y','quarter_y','year_x','month_x','quarter_x','match_period'],axis=1)
data2
```

```python
data3 = data2.fillna(data2.median())

data3['*year'] = data3['mdate'].astype(str).str[0:4]
data3['*year'] = pd.to_numeric(data3['*year'])


data3 = data3._get_numeric_data()


data3 = data3.dropna(axis='columns')


Train = data3[data3['*year'] < 2016]
Test = data3[data3['*year'] >= 2016]

Train_Y = Train['spread5y']
Train_X = Train.drop('spread5y', axis=1)

Test_Y = Test['spread5y']
Test_X = Test.drop('spread5y', axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 50)
forest_fit = rf.fit(Train_X, Train_Y)
predict_y = rf.predict(Test_X) 

print("Mean accuracy: ", forest_fit.score(Test_X, Test_Y))
```
